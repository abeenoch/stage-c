```python
x=df.iloc[:,:12]
y=df.iloc[:,13]
from sklearn.model_selection import train_test_split
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2,random_state=1)
```

```python
from sklearn.preprocessing import StandardScaler
sc=StandardScaler()
x_train=sc.fit_transform(x_train)
x_test=sc.transform(x_test)
```

```python
#import random forest model
import sklearn.linear_model as sk
from sklearn.ensemble import RandomForestClassifier
```

```python
#create a gaussaian classifier 
clf=RandomForestClassifier(n_estimators=100)
#train the model using the training sets y_pred=clf.predict(x_test)
clf.fit(x_train,y_train)
y_pred=clf.predict(x_test)
```

```python
#import scikit-learn metrics module for accuracy calculation
from sklearn import metrics
#Model Accuracy,how often is the classifier correct?
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.tree import DecisionTreeClassifier #import Decision Tree Classifier
#import train_test_split function
 #import scikit_learn metrics module for accuraccy calculation
clf=DecisionTreeClassifier()#create decisiontreeclassifierobject
clf=clf.fit(x_train,y_train)#traindecisiontreeclassifier
#predict the response for test dataset
y_pred = clf.predict(x_test)
```

```python
#Evaluating model
#model accuracy
print("Accuracy:",metrics.accuracy_score(y_test,y_pred))
```

```python
from sklearn.ensemble import RandomForestClassifier
#set random State
rf = RandomForestClassifier(random_state=1)
#import randomizedsearchcv
from sklearn.model_selection import RandomizedSearchCV
import random
```

```python
#provide hyperparameter grid for a random search
n_estimators =[50,100,300,500,1000]
max_features =['auto','sqrt','log2',None]
min_samples_split = [2,3,5,7,9]
min_samples_leaf = [1,2,4,6,8]

hyperparameter_grid     =     {'n_estimators':n_estimators,
                              'min_samples_split':min_samples_split,
                              'min_samples_leaf':min_samples_leaf,
                              'max_features':max_features}
rf_random = RandomizedSearchCV(estimator=rf,param_distributions = hyperparameter_grid,scoring='accuracy',n_iter=10,cv=5,verbose=1,random_state=1,n_jobs = -1)
rf_random.fit(x,y)
print('Hyperparameter grid',hyperparameter_grid,'\n')
print('Best Parameters:',rf_random.best_params_,'\n')
```
